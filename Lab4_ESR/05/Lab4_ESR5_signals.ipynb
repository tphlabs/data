```python
import numpy as np
import matplotlib.pyplot as plt
from uncertainties import ufloat
```

# Variable capacitor at 25 out of 100

## 5 ms scale

```python
url = 'https://raw.githubusercontent.com/tphlabs/data/refs/heads/main/Lab4_ESR/05'
folder = '/capacitor_at_25'
for ch in [1, 2]:
  fname = f'/5ms/F0001CH{ch}.CSV'
  t, v = np.genfromtxt(url + folder + fname, usecols=[3,4], unpack=True, delimiter=',')
  plt.plot(t*1e3, v, label=f'CH{ch}')
plt.legend()
plt.xlabel('Time (ms)')
plt.ylabel('Voltage (V)')

plt.grid()
```

## 10ns scale

```python
url = 'https://raw.githubusercontent.com/tphlabs/data/refs/heads/main/Lab4_ESR/05'
folder = '/capacitor_at_25'
fname = f'/10ns/F0000CH2.CSV'
t, v = np.genfromtxt(url + folder + fname, usecols=[3,4], unpack=True, delimiter=',')
v -= np.mean(v)
# find ix of sign change from plus to minus, i.e crossing zero line
z = np.sign(v[:-1]) > np.sign(v[1:])
t, v = t[1:], v[1:]
plt.plot(t[z]*1e9, v[z], 'x')
plt.plot(t*1e9, v, '-')
plt.xlabel('Time (ns)')
plt.ylabel('Voltage (V)')
plt.grid()
freq = 1e-6 / np.diff(t[z])
ufreq = ufloat(np.mean(freq), np.std(freq))
print(f'Frequency {ufreq} MHz')
```
